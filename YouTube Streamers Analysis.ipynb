## Categories with exceptional performance metrics:

```python
#Calculate average subscribers per category
average_subscribers_per_category =df.groupby('Categories')['Suscribers'].mean()

#Sort the average subcribers for each category in ascending order
average_subscribers_per_category_sorted = average_subscribers_per_category.sort_values()

# Display the average subscribers for each category in ascening order
print(average_subscribers_per_category_sorted)
```

### Average Subscribers per Category:

```python
#Create a bar plot for average subscribers per category in ascending order
plt.figure(figsize=(10, 8))

average_subscribers_per_category_sorted.plot(kind='bar', color='skyblue')
plt.title('Average Subscribers per categories', color = 'purple',fontweight='bold',fontsize=18)
plt.xlabel('Categories', color = 'red',fontweight='bold',fontsize=17)
plt.ylabel('Average Subscribers', color = 'red',fontweight='bold',fontsize=17)

plt.tight_layout()

plt.show()
```

### Categories with Exceptional Performance Metrics:

```python
plt.figure(figsize = (12,8))

metric = df['Suscribers'].mean()
b = df.groupby('Categories')['Suscribers'].mean().sort_values(ascending = False)

colors = ['pink','gold','skyblue','blue']
b = b[b > metric].plot(kind = 'bar', color = colors)

b.set_ylabel('Average Subscibers', color='brown',fontweight='bold', fontsize=16)
b.set_xlabel('Categories',color='brown',fontweight='bold', fontsize=16)
b.set_title('Categories with Exceptional Performance Metrics', color='purple', fontweight='bold', fontsize=17)
plt.show()
```

### Average Subscribers By Categories:

### Above Average Subcribers by Categories:

```python
a = df[['Categories', 'Suscribers']]
b = a.groupby('Categories')['Suscribers'].mean().reset_index()
above_avg_subscribers = b[b['Suscribers']>average_metrics['Suscribers']][['Categories','Suscribers']]
above_avg_subscribers
```

### Above Average Visits by Categories:

# CONTENT RECOMMENDATIONS:

#### Content recommendations based on streamers categories and performance metrics:

```python
recommendations = df.groupby('Categories')['Categories'].count().sort_values(ascending=False).head(10)
recommendations = pd.DataFrame({'Category_Count': recommendations})
recommendations.index.name = 'Categories'
content_recommendations = recommendations.reset_index()
content_recommendations
```

```python

plt.figure(figsize=(10, 8))
plt.barh(content_recommendations['Categories'],content_recommendations['Category_Count'],color=['pink','blue','skyblue'])

plt.title('Content Reecommendations',color = 'purple',fontweight='bold',fontsize=18)
plt.xlabel('Category_Count', color = 'red',fontweight='bold',fontsize=17)
plt.ylabel('Categories', color = 'red',fontweight='bold',fontsize=17)
plt.show()
```
